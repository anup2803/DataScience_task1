#  Hypothesis Formulation and Statistical Testing


# Research Question
 Do waterfront houses have significantly higher prices than non-waterfront houses?

# Null Hypothesis (H₀)
There is no significant difference in average house prices between waterfront and non-waterfront properties.

# Alternative Hypothesis (H₁)
Waterfront properties have significantly different average prices than non-waterfront properties.

# Correlation Test


from scipy.stats import pearsonr

It evaluates how strongly two continuous variables are connected via a linear relationship



```python
from scipy.stats import ttest_ind

waterfront = df[df['waterfront'] == 1]['price']
non_waterfront = df[df['waterfront'] == 0]['price']

t_stat, p_value = ttest_ind(waterfront, non_waterfront)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

# Interpretation


T-statistic = 40.62
P-value = 0.0

The p-value is less than the significance level of 0.05.

Therefore, the null hypothesis is rejected.

There is a statistically significant difference in house prices between waterfront and non-waterfront properties.

Waterfront houses tend to have much higher prices than non-waterfront houses in King County.

# Conclusion
An independent two-sample t-test was conducted to determine whether waterfront properties have different average prices compared to non-waterfront properties. The test produced a t-statistic of 40.62 and a p-value of 0.0. Since the p-value is less than the significance level of 0.05, the null hypothesis was rejected. Therefore, there is strong statistical evidence that waterfront houses are significantly more expensive than non-waterfront houses in King County.
